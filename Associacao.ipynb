```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
import itertools
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://raw.githubusercontent.com/marcellaresende/AssociationRules/main/BreadBasket_DMS.csv'

df = pd.read_csv(url)
```

```python
transactions = df[df['Item'] != 'NONE'].groupby('Transaction')['Item'].apply(list)


te = TransactionEncoder()
data_encoded = te.fit_transform(transactions)
df_encoded = pd.DataFrame(data_encoded, columns=te.columns_)
```

```python
df_encoded.head()
```

```python
min_support = 0.01
frequent_itemsets_apriori = apriori(df_encoded, min_support=min_support, use_colnames=True, verbose = 1)
frequent_itemsets_apriori.sort_values(by = "support", ascending = False)
```

```python
rules = association_rules(frequent_itemsets_apriori, metric= "confidence", min_threshold = 0.5)

rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])

rules
```

A diferença nesse caso ficou quanto a abordagem técnica dos algoritmos, uma vez que as regras geradas foram muito semelhantes.

**b) Para ambos os algoritmos, realize testes com 4 diferentes valores de suporte mínimo e confiança mínima. O que acontece? Por que?**

Em resumo, quando alteramos os valores de suporte e confiança isso impacta diretamente na quantidade e qualidade das regras geradas pelos algoritmos.

```python
min_support = [0.01, 0.02, 0.03, 0.04]
min_confidence = [0.5, 0.6, 0.7, 0.8]
```

```python
for support, confidence in itertools.product(min_support, min_confidence):
    frequent_itemsets_apriori = apriori(df_encoded, min_support=support, use_colnames=True, verbose=0)
    rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=confidence)
    print(f"Apriori - Support: {support}, Confidence: {confidence}")
    print(rules_apriori)
    print("-----------------------------------------")
```

```python
for support, confidence in itertools.product(min_support, min_confidence):
    frequent_itemsets_fpgrowth = fpgrowth(df_encoded, min_support=support, use_colnames=True, verbose=0)
    rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=confidence)
    print(f"FP-Growth - Support: {support}, Confidence: {confidence}")
    print(rules_fpgrowth)
    print("-----------------------------------------")
```
